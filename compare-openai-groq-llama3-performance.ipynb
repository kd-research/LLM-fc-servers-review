```python
import os
from autogen import runtime_logging
from autogen import AssistantAgent, UserProxyAgent, GroupChat, GroupChatManager
```

```python
llm_config = {"model": "llama3-8b-8192", "api_key": os.environ["GROQ_API_KEY"], "base_url": os.environ["GROQ_BASE_URL"], "cache_seed": None}
llm_config = {"model": "gpt-4o", "api_key": os.environ["OPENAI_API_KEY"], "cache_seed": None}
```

```python
logging_session_id = runtime_logging.start(config={"dbname": "logs.db"})
print("Logging session ID: " + str(logging_session_id))

assistants = []

player1 = AssistantAgent(
    f"player_1", 
    llm_config=llm_config,
    system_message=f"""You are a player in a game to guess a number between 1 to 100. 
    Please win this game with your divide-and-conqure skills.
    In order to submit your guessed number, call function 'submit' with your number eg submit(10)
    The function will return low if your guessed number is lower that the target, high otherwise.
    When you see a 'bingo' from any player, you should reply "TERMINATE" to exit the game.
    """,
)

player2 = AssistantAgent(
    f"player_2", 
    llm_config=llm_config,
    system_message=f"""You are a player in a game to guess a number between 1 to 100. 
    Please win this game with your divide-and-conqure skills.
    In order to submit your guessed number, call function 'submit' with your number eg submit(10)
    The function will return low if your guessed number is lower that the target, high otherwise.
    When you see a 'bingo' from any player, you should reply "TERMINATE" to exit the game.
    """,
)

user_proxy = UserProxyAgent(
    "game_host", 
    is_termination_msg=lambda x: (x is not None) and (x["content"] is not None) and ("TERMINATE" in x.get("content", "TERMINATE")), 
    human_input_mode="NEVER",
    max_consecutive_auto_reply=20,
    llm_config=llm_config,

)
last=None
def state_transition(last_speaker, _):
    global last
    if last_speaker is user_proxy:
        if last is player1:
            last = player2
            return player2
        last = player1
        return player1
    return user_proxy
    
groupchat = GroupChat(
    agents=[user_proxy, player1, player2],
    messages=[],
    max_round=30,
    speaker_selection_method=state_transition,
)
manager = GroupChatManager(groupchat=groupchat, llm_config=llm_config)

@user_proxy.register_for_execution()
@player1.register_for_llm(description="This is where you put your number in")
@player2.register_for_llm(description="This is where you put your number in")
def submit(num: int) -> str:
    if num < 19:
        return "low"
    elif num > 19:
        return "high"
    return "bingo"

user_proxy.initiate_chat(manager, message="Let's begin this game. The target number is already in my mind. Please pick a random number with submit(your-number).")

autogen.runtime_logging.stop()
```
